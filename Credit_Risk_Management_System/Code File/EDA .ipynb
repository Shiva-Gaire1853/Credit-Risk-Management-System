```python
#now collecting all the columns which have more than 10000 rows null value
col_with_null_values= []
for i in df2.columns:
    if (df2.loc[df2[i]== -99999].shape[0]>10000):
        col_with_null_values.append(i)
```

```python
print(col_with_null_values)
```

```python
#since this column contain very high amount of null value so we will gone a remove it
df2 = df2.drop(columns=col_with_null_values,axis=1)
```

```python
#now will we will see other null values and if it is in less amount than we will remove it
for i in df2.columns:
    df2 = df2.loc[df2[i]!=-99999]
```
